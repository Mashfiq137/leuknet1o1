```python
PATH = r'../input/cmncthesis'
PATH
```

```python
paths_para=glob.glob(os.path.join(PATH,'C-NMC_test_final_phase_data','*.bmp'))
```

```python
def crop_center(img, bounding):
    start = tuple(map(lambda a, da: a//2-da//2, img.shape, bounding))
    end = tuple(map(operator.add, start, bounding))
    slices = tuple(map(slice, start, end))
    return img[slices]
```

```python
def get_data(path, resize_dim=210):
    img=cv2.imread(path)
    img=crop_center(img,(210,210))
    
    #img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA) 
    #img_rgb = np.asarray(np.dstack((img, img, img)), dtype=np.uint8)
    
    img_rz=cv2.resize(img,(resize_dim,resize_dim),interpolation=cv2.INTER_AREA)
    
    return img_rz
```

```python
images=np.array([get_data(path) for path in paths_para])
images.shape
```
